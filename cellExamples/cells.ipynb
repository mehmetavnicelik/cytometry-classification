```python
from numpy import asarray
img=rasterio.open("/home/avni/Desktop/gp/wspace/cellExamples/ros23ch4.tif")
# plt.imshow(img.read(1),cmap='gray')
# plt.text(10, 5, "ros23-ch4",size="large",color="black")
a=img.read(1)
print(a)
print(a.shape)
print(a.size)
```

```python
norm = cv2.normalize(a, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
print(norm.shape, norm.dtype)

# convert to 3 channel
norm = cv2.cvtColor(norm, cv2.COLOR_GRAY2BGR)
print(norm.shape, norm.dtype)
print(np.amin(norm),np.amax(norm))
```
